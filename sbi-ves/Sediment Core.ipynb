```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager
import matplotlib.patches as patches
```

```python
# Define the layers
layers = np.array([2.4, 1.0, 6.6, 8.6, 2.1])
layeres_reshaped = layers.reshape(5,1)
cumulative_depth = np.cumsum(layers)
colors = ['#c1a25fff', '#fdfe9aff', '#fdc800ff', '#fdfe9aff', '#949393ff']  # Similar colors for the 2nd and 4th layers

# Define the y positions (top edges of each layer)
y_positions = [0,2.4,3.4,10,18.6, 20.7]

with mpl.rc_context(fname="plots/pltstyle.rc"):
    # Create the figure and axis
    fig, ax1 = plt.subplots(figsize=(6, 10))
    
    
    # Plot each layer as a bar
    for i, (depth, cdepth) in enumerate(zip(layers, y_positions)):
        ax1.bar(x=[1],y=y_positions[i], width=0.2, height=depth, color=colors[i], edgecolor='black')
    
    # Set the labels for the depth transitions
    for depth in cumulative_depth:
        ax1.text(1.1, depth-0.25, '___{:.1f}'.format(depth), va='center', ha='left', fontweight='bold')


    # First layer details
    first_layer_details = "slit/sand \n{} - {}$\Omega m$".format(50,250)
    plt.annotate(first_layer_details, xy=(0.3, 1.5), fontweight='bold')
    
    # Second Layer details
    second_layer_type = 'gravel\n'
    plt.annotate(second_layer_type, xy=(0.3, 3.5), fontweight='bold')

    second_layer_detail = '\n(slity/sandy)'
    plt.annotate(second_layer_detail, xy=(0.3,3.5), fontsize=12)

    second_layer_res = '\n \n{} - {}$\Omega m$'.format(100,500)
    plt.annotate(second_layer_res, xy=(0.3,4), fontweight='bold')
    
    # Third layer details
    third_layer_type = 'sand'
    plt.annotate(third_layer_type, xy=(0.3, 7.5), fontweight='bold')
    third_layer_detail = '(fine sand/gravely)'
    plt.annotate(third_layer_detail, xy=(0.3,8), fontsize=12)
    third_layer_res = '\n \n \n{} - {}$\Omega m$'.format(50,250)
    plt.annotate(third_layer_res, xy=(0.3,8.5), fontweight='bold')

    
    
    # Fourth layer details
    fourth_layer_type = 'gravel'
    plt.annotate(fourth_layer_type, xy=(0.3, 15), fontweight='bold')
    fourth_layer_detail = '(slity/sandy)'
    plt.annotate(fourth_layer_detail, xy=(0.3,15.5), fontsize=12)
    fourth_layer_res = '\n \n \n{} - {}$\Omega m$'.format(100,500)
    plt.annotate(fourth_layer_res, xy=(0.3, 16), fontweight='bold')

    
    # Fifth layer details
    fifth_layer_details = "slit \n{} - {}$\Omega m$".format(10,70)
    plt.annotate(fifth_layer_details, xy=(0.3,20), fontweight='bold')
    
    
    
    # Set the limits and labels
    ax1.set_xlim(0, 1.5)
    ax1.set_ylim(0, cumulative_depth[-1])
    ax1.set_xticks([])  # Remove x-axis ticks
    ax1.set_yticks([])
    #ax1.set_title('Drilling Core Depth Profile')
    ax1.set_ylabel('Depth (m)')
    ax1.spines['bottom'].set_visible(False)
    ax1.spines['left'].set_visible(False)
    
    # Invert the y-axis so that depth increases downwards
    ax1.invert_yaxis()
    
    # Show the plot
    plt.savefig('plots/drilling core/drilling_core.svg')
    plt.show()
```
